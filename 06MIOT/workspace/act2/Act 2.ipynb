```python
import pandas as pd
import numpy as np
```

### Carga de datos de csv(estaciones_bici.csv) en un DataFrame

```python
# Read CSV
df_stations = pd.read_csv('./examples/estaciones_bici.csv', delimiter=';')
```

```python
df_stations
```

### Obtención de número de estaciones con total de 30

```python
df_stations_with_30_spots = df_stations[df_stations.total == 30]
```

```python
df_stations_with_30_spots.groupby('estation')["total"].aggregate([max])
```

### Número de estación con la media más alta de bicis disponibles

```python
df_with_mean_calculated = df_stations.groupby('estation')["available"].aggregate([np.mean])
```

```python
df_with_mean_calculated
```

La estación con la **media más alta** de bicicletas disponibles es la número **50**:

```python
df_with_mean_calculated.loc[df_with_mean_calculated.idxmax()]
```

## Partiendo de esta estación ##

### Realizar el histograma de la estación de bicis disponibles ###

```python
# Uso de la libreria matplotlib

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df_station_50 = df_stations[df_stations.estation == 50]
df_station_50
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(df_station_50.available)
```

### Realizar gráfica con la línea temporal de bicis disponibles ###

Si nos fijamos en la columna **'updated'**, tenemos el dato de actualización de la estación, por lo que podemos utilizar este valor para nuestra línea temporal:

```python
fig, ax = plt.subplots(figsize=(15, 15))

# add the dates axis and the values
ax.plot(df_station_50.updated, df_station_50.available)

# rotate and align the tick labels so they look better
fig.autofmt_xdate()

# give more room printing the tick every 4 points
plt.xticks(np.arange(0, 100, step=4))

# print the plot
plt.show()
```
